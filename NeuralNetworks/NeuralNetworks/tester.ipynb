```python
from network import *
from numpy import array
```

```python
X = array([
    [0,0],
    [0,1],
    [1,0],
    [1,1]
])

Y = array([
    [1,0],
    [0,1],
    [0,1],
    [1,0]
]) 
```

```python
model = Network(
    layerSizes=[2,4,2],
    activations=["relu","softmax"],
    lossName="crossentropy",
    optimizer="sgd"
)

model.Train(X, Y, learningRate=0.1, epochs=1000, batchSize=4)

model.TrainSummary(epochsPerPrint=100)
```
