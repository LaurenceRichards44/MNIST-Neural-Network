```python
#from NeuralNetworks import *
import network
from torchvision.datasets import MNIST
from numpy import concatenate, eye, int32, float32
```

```python
train_dataset = MNIST(root="data/", train=True, download=True)
test_dataset = MNIST(root="data/", train=False, download=True)

X_train = train_dataset.data.numpy().astype(float32).reshape(len(train_dataset), -1) / 255.0
Y_train = eye(10)[train_dataset.targets.numpy().astype(int32)]

X_test = test_dataset.data.numpy().astype(float32).reshape(len(test_dataset), -1) / 255.0
Y_test = eye(10)[test_dataset.targets.numpy().astype(int32)]

X = concatenate((X_train, X_test), axis=0)
Y = concatenate((Y_train, Y_test), axis=0)
```
